# Non-linear constrained Gaussian random field code

Python libraries

```python
import numpy as np
import scipy.integrate as integrate
import h5py
import matplotlib.pyplot as plt
```

Cosmological setup

```python
def normalize(L):
    Theta = 2.7255 / 2.7
    R = 8.
    sigma = 0.1
    Omega0 = 1.
    h = 0.674
    ns = 0.965
    sigma8 = 0.811
    
    def q(k):
        return k * Theta ** 2 / (Omega0 * h)
    def C0(q):
        return 14.2 + 731. / (1. + 62.5 * q)
    def L0(q):
        return np.log(2. * np.exp(1) + 1.8 * q)
    def T0(q):
        return L0(q) /(L0(q) + C0(q) * q ** 2)
    def W(y):
        return 3. / y**3 * (np.sin(y) - y * np.cos(y))
    
    x = integrate.quad(lambda k: 
                       1. / (2. * np.pi ** 2.) * k ** ns * T0(q(k)) ** 2. * W(k * R) ** 2. * k ** 2., 
                       2. * np.pi / L, np.inf)

    return  sigma8 ** 2 / x[0];

def P(kx, ky, kz, L):
    Theta = 2.7255 / 2.7
    R = 8.
    sigma = 0.1
    Omega0 = 1.
    h = 0.674
    ns = 0.965
    sigma8 = 0.811


    def q(k):
        return k * Theta ** 2 / (Omega0 * h)
    def C0(q):
        return 14.2 + 731. / (1. + 62.5 * q)
    def L0(q):
        return np.log(2. * np.exp(1) + 1.8 * q)
    def T0(q):
        return L0(q) /(L0(q) + C0(q) * q**2)
    
    np.seterr(divide='ignore')
    Pk = (np.sqrt(kx**2 + ky**2 + kz**2) ** (-4. + ns) * T0(q(np.sqrt(kx**2 + ky**2 + kz**2)))**2 *   
          np.exp(-(sigma * np.sqrt(kx**2 + ky**2 + kz**2)) ** 2.))
    Pk[0,0,0] = 0
    return Pk
```

Constraint Gaussian random field functions

```python
def kRange(n, L):
    return np.fft.fftfreq(n) * n * 2 * np.pi / L

def convolve(a, b):
    return np.real(np.fft.ifftn(np.conjugate(a) * np.fft.fftn(b)))

def mean(c, bases):
    return np.einsum("ijkl, i -> jkl", bases, c)

def measureConstraints(f, Hh):
    return np.array([np.real(np.mean(np.conjugate(h) * np.fft.fftn(f))) for h in Hh])

def GRF(Pk, Ni, L, seed):
    np.random.seed(seed)
    return (np.real(np.fft.ifftn(np.sqrt(Pk) * np.fft.fftn(np.random.normal(0.0, 1.0, size = (Ni, Ni, Ni))))) 
            * Ni**(3/2) / L**(3/2))

def constraintGRF(c, Hh, bases, Pk, Ni, L, seed):
    np.random.seed(seed)
    f = GRF(Pk, Ni, L, seed)
    fc = f - mean(measureConstraints(f, Hh), bases) + mean(c, bases)
    return (f, fc)

def variance(xi_i, xi_ij):
    return np.einsum("iklm, ij, jklm -> klm", xi_i, np.linalg.inv(xi_ij), xi_i)

def laplace(f, kx, ky, kz):
    return convolve(-kx * kx - ky * ky - kz * kz, f)

class cGRF:
    def __init__(self, Ni, L, sigma):
        self.Ni = Ni
        self.L = L
        self.sigma = sigma
        
        self.kz = np.tile(kRange(Ni, L),((Ni * Ni, 1))).reshape(Ni, Ni, Ni)
        self.ky = self.kz.transpose((0,2,1))
        self.kx = self.kz.transpose((2,0,1))

        self.ikx = 1j * self.kx
        self.iky = 1j * self.ky
        self.ikz = 1j * self.kz
        
        self.Pk = normalize(L) * P(self.kx, self.ky, self.kz, self.L)
        self.sigmaK = np.exp(- sigma**2 * (self.kx**2 + self.ky**2 + self.kz**2) / 2.)

        (xd, yd, zd) = (L / 2., L / 2., L / 2.)
        expIKX = np.exp(1j * (self.kx * xd + self.ky * yd + self.kz * zd))
        
        self.Hh = (self.sigmaK * expIKX * self.ikx * self.ikx, 
                   self.sigmaK * expIKX * self.ikx * self.iky, 
                   self.sigmaK * expIKX * self.ikx * self.ikz, 
                   self.sigmaK * expIKX * self.iky * self.iky, 
                   self.sigmaK * expIKX * self.iky * self.ikz,
                   self.sigmaK * expIKX * self.ikz * self.ikz, 

                   self.sigmaK * expIKX * self.ikx * self.ikx * self.ikx, 
                   self.sigmaK * expIKX * self.ikx * self.ikx * self.iky, 
                   self.sigmaK * expIKX * self.ikx * self.ikx * self.ikz, 
                   self.sigmaK * expIKX * self.ikx * self.iky * self.iky,
                   self.sigmaK * expIKX * self.ikx * self.iky * self.ikz,
                   self.sigmaK * expIKX * self.ikx * self.ikz * self.ikz,
                   self.sigmaK * expIKX * self.iky * self.iky * self.iky,
                   self.sigmaK * expIKX * self.iky * self.iky * self.ikz,
                   self.sigmaK * expIKX * self.iky * self.ikz * self.ikz,
                   self.sigmaK * expIKX * self.ikz * self.ikz * self.ikz,

                   self.sigmaK * expIKX * self.ikx * self.ikx * self.ikx * self.ikx,
                   self.sigmaK * expIKX * self.ikx * self.ikx * self.ikx * self.iky,      
                   self.sigmaK * expIKX * self.ikx * self.ikx * self.ikx * self.ikz,
                   self.sigmaK * expIKX * self.ikx * self.ikx * self.iky * self.iky,
                   self.sigmaK * expIKX * self.ikx * self.ikx * self.iky * self.ikz,
                   self.sigmaK * expIKX * self.ikx * self.ikx * self.ikz * self.ikz,
                   self.sigmaK * expIKX * self.ikx * self.iky * self.iky * self.iky,
                   self.sigmaK * expIKX * self.ikx * self.iky * self.iky * self.ikz,
                   self.sigmaK * expIKX * self.ikx * self.iky * self.ikz * self.ikz,
                   self.sigmaK * expIKX * self.ikx * self.ikz * self.ikz * self.ikz,
                   self.sigmaK * expIKX * self.iky * self.iky * self.iky * self.iky,
                   self.sigmaK * expIKX * self.iky * self.iky * self.iky * self.ikz,
                   self.sigmaK * expIKX * self.iky * self.iky * self.ikz * self.ikz,
                   self.sigmaK * expIKX * self.iky * self.ikz * self.ikz * self.ikz,
                   self.sigmaK * expIKX * self.ikz * self.ikz * self.ikz * self.ikz)

        self.xi_i = np.real(np.array([np.fft.ifftn(hi * self.Pk) for hi in self.Hh])) * L**-3 * Ni**3
        self.xi_ij = np.real(np.array([np.mean(np.conjugate(hi) * hj * self.Pk)
                                  for hi in self.Hh for hj in self.Hh]).reshape(len(self.Hh),len(self.Hh))) * L**-3 * Ni**3
        self.bases = np.einsum("ij, jklm -> iklm", np.linalg.inv(self.xi_ij), self.xi_i)
    
    def mean(self, c):
        return mean(c, self.bases)
    def var(self, c):
        return variance(self.xi_i, self.xi_ij)
    def realization(self, c, seed):
        (f, fc) = constraintGRF(c, self.Hh, self.bases, self.Pk, self.Ni, self.L, seed)
        return (f, fc)
```

Setup constrained Gaussian random field generator

```python
Ni = 2**7
L = 100
sigma = 2
DpC = 0.8

generator = cGRF(Ni, L, sigma)
```

Generate constrained initial conditions

```python
ICs_file = h5py.File('ICs_A3.h5', 'r')
ICs = np.array(ICs_file['ICs_s=' + str(sigma) + '_DpC=' + str(DpC)])

c = ICs[1]
fb = generator.mean(c)
fv = generator.var(c)

(f, fc) = generator.realization(c, 1123)
```

Plot random field realization, mean field, and variance

```python
plt.figure(figsize=(10,10))
plt.matshow(np.transpose(fc[:,-1:0:-1,int(Ni/2)]), fignum=1)
plt.colorbar()
plt.show()
```
